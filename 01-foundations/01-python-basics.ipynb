```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("NumPy:", np.__version__)
print("Pandas:", pd.__version__)
print("Setup complete!")
```

```python
# A small dataset of UK city average house prices (£)
data = {
    'city': ['London', 'Manchester', 'Birmingham', 'Leeds', 'Salford'],
    'avg_price': [531000, 243000, 231000, 216000, 198000],
    'bedrooms': [2, 3, 3, 3, 3]
}
df = pd.DataFrame(data)
print(df)
```

```python
plt.figure(figsize=(8, 4))
plt.bar(df['city'], df['avg_price'], color='steelblue')
plt.title('Average house prices by UK city')
plt.ylabel('Price (£)')
plt.tight_layout()
plt.show()
```
